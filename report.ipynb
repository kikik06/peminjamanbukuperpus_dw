```python
import pandas as pd

# read all data
dataset = pd.read_csv('Library_Usage.csv')
dataset
```

```python
# print all column name and it's data type
dataset.dtypes
```

```python
# report 1
library_branch_registration = dataset['Home Library Definition'].value_counts().reset_index()
library_branch_registration.columns = ['library_branch', 'count']
library_branch_registration
```

```python
# report 1 visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(24,8))
plt.barh(library_branch_registration['library_branch'], library_branch_registration['count'])
plt.title("Library Branch of Member Registration")
plt.ylabel('Library Branch')
plt.xlabel('Count')
plt.savefig('report_1.png')
plt.show()
```

```python
# report 2
patron_type = dataset['Patron Type Definition'].value_counts().reset_index()
patron_type.columns = ['patron_type', 'count']
patron_type
```

```python
# report 2 visualization
plt.figure(figsize=(24,8))
plt.barh(patron_type['patron_type'], patron_type['count'])
plt.title("Library Member's Patron Type")
plt.ylabel('Patron Type')
plt.xlabel('Count')
plt.savefig('report_2.png')
plt.show()
```

```python
# report 3
age_range = dataset['Age Range'].value_counts().reset_index()
age_range.columns = ['age_range', 'count']
age_range = age_range.sort_values('age_range').reset_index().drop('index', 1)
age_range
```

```python
# report 3 visualization barchart horizontal
# plt.figure(figsize=(24,8))
# plt.barh(age_range['age_range'], age_range['count'])
# plt.title("Library Member's Age Range")
# plt.ylabel('Age Range')
# plt.xlabel('Count')
# plt.savefig('report_3.png')
# plt.show()

# report 3 visualization pie chart
pie_labels = age_range['age_range']
ar_count = age_range['count']
fig, ax = plt.subplots(figsize=(14,10))
ax.pie(ar_count, labels=pie_labels, autopct='%1.2f%%', startangle=90)
ax.set_title("Library Member's Age Range")
ax.axis('equal')
plt.savefig('report_3.png')
plt.show()
```

```python
# report 4
year_registered = dataset['Year Patron Registered'].value_counts().reset_index()
year_registered.columns = ['year_registered', 'count']
year_registered = year_registered.sort_values('year_registered').reset_index().drop('index', 1)
year_registered
```

```python
# report 4 visualization
plt.figure(figsize=(24,8))
plt.plot(year_registered['year_registered'], year_registered['count'], linestyle='--', marker='o')
plt.title("Library Member's Year Registered")
plt.xlabel('Year')
plt.ylabel('Count')
plt.savefig('report_4.png')
plt.show()
```

```python
# report 5
active_2015 = dataset.loc[dataset['Circulation Active Year'] == '2015']
month_active_2015 = active_2015['Circulation Active Month'].value_counts().reset_index()
month_active_2015.columns = ['month', 'count']
month_active_2015['month_number'] = pd.to_datetime(month_active_2015.month, format='%B').dt.month
month_active_2015 = month_active_2015.sort_values('month_number').reset_index().drop('index', 1)
month_active_2015
```

```python
# report 5 visualization
plt.figure(figsize=(24,8))
plt.plot(month_active_2015['month'], month_active_2015['count'], linestyle='--', marker='o')
plt.title("Library Member's Last Activity in 2015")
plt.xlabel('Month')
plt.ylabel('Count')
plt.savefig('report_5.png')
plt.show()
```

```python
# report 6
notice_preference = dataset['Notice Preference Definition'].value_counts().reset_index()
notice_preference.columns = ['notice_preference', 'count']
notice_preference = notice_preference.sort_values('notice_preference').reset_index().drop('index', 1)
notice_preference
```

```python
# report 6 visualization pie chart
pie_labels = notice_preference['notice_preference']
ar_count = notice_preference['count']
fig, ax = plt.subplots(figsize=(14,10))
ax.pie(ar_count, labels=pie_labels, autopct='%1.4f%%', startangle=90)
ax.set_title("Library Member's Notice Preference")
ax.axis('equal')
plt.savefig('report_6.png')
plt.show()
```

```python
# report 7
avg_co_by_age = dataset.groupby(['Age Range'])['Total Checkouts'].mean().reset_index()
avg_co_by_age
```

```python
# report 7 visualization
plt.figure(figsize=(24,8))
plt.barh(avg_co_by_age['Age Range'], avg_co_by_age['Total Checkouts'])
plt.title("Library Member's Average Checkouts Based on Age Range")
plt.ylabel('Age Range')
plt.xlabel('Average Checkouts')
plt.savefig('report_7.png')
plt.show()
```

```python
# report 8
avg_renewals_by_pt = dataset.groupby(['Patron Type Definition'])['Total Renewals'].mean().reset_index()
avg_renewals_by_pt
```

```python
# report 8 visualization
plt.figure(figsize=(24,8))
plt.barh(avg_renewals_by_pt['Patron Type Definition'], avg_renewals_by_pt['Total Renewals'])
plt.title("Library Member's Average Renewals Based on Patron Type")
plt.ylabel('Patron Type')
plt.xlabel('Average Renewals')
plt.savefig('report_8.png')
plt.show()
```
